```python
# we have to drop cols like car_id
# we have to make a single column named condition with the help of core_rating, supporting_rating, interior_rating, exterior_rating, wear_rating
# may be some columns need to renamed like cylinder has to renamed as cylinders and types as type
```

```python
# we have to change the type of the cylinder col to int
```

```python
df.drop(columns = ['car_id'], inplace = True)
```

```python
df.rename(columns = {'cylinder': 'cylinders', 'types': 'type'}, inplace = True)
```

```python
df.head()
```

```python
# need to change new i20 to i20
# values of region, manufacturer, model, drive has to be in lower case
```

```python
df['region'] = df['region'].str.lower()
df['manufacturer'] = df['manufacturer'].str.lower()
df['model'] = df['model'].str.lower()
df['drive'] = df['drive'].str.lower()
df['fuel'] = df['fuel'].str.lower()
df['transmission'] = df['transmission'].str.lower()
df['type'] = df['type'].str.lower()
```

```python
df.drop(columns = (['core_rating', 'supporting_rating', 'interior_rating', 'exterior_rating', 'wear_rating']), inplace = True)
```
