```python
df.loc[df['model'] == '3 series', 'type'] = df.loc[df['model'] == '3 series', 'type'].fillna('sedan')
```

```python
df.loc[df['model'] == 'xc90', 'type'] = df.loc[df['model'] == 'xc90', 'type'].fillna('suv')
```

```python
df.loc[df['model'] == 's 80', 'type'] = df.loc[df['model'] == 's 80', 'type'].fillna('sedan')
```

```python
df.loc[df['model'] == 's-class', 'type'] = df.loc[df['model'] == 's-class', 'type'].fillna('sedan')
```

```python
df.loc[df['model'] == 'huracan evo', 'type'] = df.loc[df['model'] == 'huracan evo', 'type'].fillna('coupe')
```

```python
df.loc[df['model'] == 'gtc4lusso', 'type'] = df.loc[df['model'] == 'gtc4lusso', 'type'].fillna('coupe')
```

```python
df.loc[df['model'] == '5 series', 'type'] = df.loc[df['model'] == '5 series', 'type'].fillna('sedan')
```

```python
df[(df['model'] == 'camry') & (df['engine_cc'] == 0)]
```

```python
df.loc[df['model'] == 'camry', 'cylinders'] = df.loc[df['model'] == 'camry', 'cylinders'].replace(0, 4)
```

```python
df.loc[df['model'] == 'camry', 'fuel'] = df.loc[df['model'] == 'camry', 'fuel'].replace('electric', 'petrol')
```

```python
df.loc[
    (df['engine_cc'] == 0) & (df['fuel'] == 'petrol'),
    'engine_cc'
] = np.nan
```

```python
df.isnull().sum()
```

```python
df['engine_cc'] = df.groupby(
    ['manufacturer', 'model', 'max_power']
)['engine_cc'].transform(
    lambda x: x.fillna(x.median() if not np.isnan(x.median()) else df['engine_cc'].median())
)
```
