```python
df['price'] = df['price'].astype(str).str.replace(',', '')
```

```python
def clean_price(x):

    x = float(x)

    if x > 1000:   # price written in rupees
        return x / 100000

    else:          # already in lakhs
        return x
```

```python
df['price'] = df['price'].apply(clean_price)
```

```python
df.head()
```

```python
df['price'] = df['price'].astype(float).astype(int)
```

```python
threshold = df['price'].quantile(0.98)

df = df[df['price'] <= threshold]
```
